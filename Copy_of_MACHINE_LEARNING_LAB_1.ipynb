Objective-2 :
To practice selecting columns and filtering rows using Pandas.

Tasks:

Objective-3 :
To perform numerical computations on student scores using NumPy.

Tasks:
1. Convert score columns into NumPy arrays.
2. Compute mean, median, standard deviation.
3. Find minimum and maximum scores.

```python
import numpy as np

math_scores_np = df['math_score'].to_numpy()
science_scores_np = df['science_score'].to_numpy()
english_scores_np = df['english_score'].to_numpy()
overall_scores_np = df['overall_score'].to_numpy()

print("Math scores as NumPy array (first 5 values):")
display(math_scores_np[:5])
print("Type of math_scores_np:", type(math_scores_np))
```

```python
print("Math Scores:")
print(f"  Mean: {np.mean(math_scores_np):.2f}")
print(f"  Median: {np.median(math_scores_np):.2f}")
print(f"  Standard Deviation: {np.std(math_scores_np):.2f}")

print("\nScience Scores:")
print(f"  Mean: {np.mean(science_scores_np):.2f}")
print(f"  Median: {np.median(science_scores_np):.2f}")
print(f"  Standard Deviation: {np.std(science_scores_np):.2f}")

print("\nEnglish Scores:")
print(f"  Mean: {np.mean(english_scores_np):.2f}")
print(f"  Median: {np.median(english_scores_np):.2f}")
print(f"  Standard Deviation: {np.std(english_scores_np):.2f}")

print("\nOverall Scores:")
print(f"  Mean: {np.mean(overall_scores_np):.2f}")
print(f"  Median: {np.median(overall_scores_np):.2f}")
print(f"  Standard Deviation: {np.std(overall_scores_np):.2f}")
```

```python
print("Math Scores:")
print(f"  Minimum: {np.min(math_scores_np):.2f}")
print(f"  Maximum: {np.max(math_scores_np):.2f}")

print("\nScience Scores:")
print(f"  Minimum: {np.min(science_scores_np):.2f}")
print(f"  Maximum: {np.max(science_scores_np):.2f}")

print("\nEnglish Scores:")
print(f"  Minimum: {np.min(english_scores_np):.2f}")
print(f"  Maximum: {np.max(english_scores_np):.2f}")

print("\nOverall Scores:")
print(f"  Minimum: {np.min(overall_scores_np):.2f}")
print(f"  Maximum: {np.max(overall_scores_np):.2f}")
```

Objective-4 :
To visualize student performance using Matplotlib.

Tasks:
1. Plot histogram of math scores.
2. Plot bar chart of average scores.
3. Create scatter plot between math and writing scores.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(math_scores_np, bins=20, edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

average_math = np.mean(math_scores_np)
average_science = np.mean(science_scores_np)
average_english = np.mean(english_scores_np)
average_overall = np.mean(overall_scores_np)

average_scores = {
    'Math': average_math,
    'Science': average_science,
    'English': average_english,
    'Overall': average_overall
}

score_names = list(average_scores.keys())
score_values = list(average_scores.values())

plt.figure(figsize=(10, 6))
plt.bar(score_names, score_values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Average Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100) # Scores are typically out of 100
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(math_scores_np, english_scores_np, alpha=0.5)
plt.title('Scatter Plot of Math Scores vs. English Scores')
plt.xlabel('Math Score')
plt.ylabel('English Score')
plt.grid(True)
plt.show()
```

Objective-5 :
To analyze student performance based on different groups.

Tasks:
1. Group scores by gender.
2. Group scores by parental education.
3. Visualize group-wise average scores using bar charts.

```python
gender_grouped_scores = df.groupby('gender')[['math_score', 'science_score', 'english_score', 'overall_score']].mean()

print("Average scores grouped by gender:")
display(gender_grouped_scores)
```

```python
parent_education_grouped_scores = df.groupby('parent_education')[['math_score', 'science_score', 'english_score', 'overall_score']].mean()

print("Average scores grouped by parental education:")
display(parent_education_grouped_scores)
```

```python
import matplotlib.pyplot as plt

# Plotting gender-wise average scores
print("Visualizing Average Scores by Gender:")
gender_grouped_scores.plot(kind='bar', figsize=(12, 7), colormap='viridis')
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend(title='Score Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting parental education-wise average scores
print("\nVisualizing Average Scores by Parental Education:")
parent_education_grouped_scores.plot(kind='bar', figsize=(12, 7), colormap='plasma')
plt.title('Average Scores by Parental Education')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend(title='Score Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
