```python
# Load the dataset using Pandas.
import pandas as pd
df = pd.read_csv('Salary_dataset.csv')
```

```python
# Display first five rows.
df.head()
```

```python
df.head()
```

```python
# Identify input and output variables.
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

Objective:
To implement Linear Regression using Scikit-learn library.

Tasks:
1. Split data into training and testing sets.
2. Import LinearRegression from sklearn.
3. Train the model using fit().
4. Predict output values.

Conclusion: Linear regression model was successfully implemented using Scikit-learn.


```python
# Split data into training and testing sets.
from sklearn.model_selection import train_test_split
```

```python
# Import LinearRegression from sklearn.
from sklearn.linear_model import LinearRegression
```

```python
# Split data into training and testing sets and train the model.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Initialize the Linear Regression model.
regressor = LinearRegression()

# Train the model using fit().
regressor.fit(X_train, y_train)
```

```python
# Predict output values.
y_pred = regressor.predict(X_test)
```

Objective:
To evaluate the performance of Scikit-learn Linear Regression model.

Tasks:
1. Calculate Mean Squared Error.
2. Calculate R-squared value.
3. Interpret model accuracy.

Conclusion: Model performance metrics were evaluated successfully.


```python
# Calculate Mean Squared Error.
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
# Calculate R-squared value
r2 = r2_score(y_test, y_pred)
print(f'R-squared (R2) Value: {r2}')
```
